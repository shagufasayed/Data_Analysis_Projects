```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)
```

```python
df=pd.read_csv('D:\Shagufa\MoviesDataAnalysis\movies.csv')
```

```python
df
```

```python
df.dtypes
```

```python
for col in df.columns:
    missing=np.mean(df[col].isnull())
    print('{} - {}%'.format(col,round(missing*100)))
    
```

```python
df = df.dropna()
df
```

```python
df['budget']=df['budget'].astype('int64')
df['gross']=df['gross'].astype('int64')
df['votes']=df['votes'].astype('int64')
```

```python
df.dtypes
```

```python
df['yearcorrect'] = df['released'].str.extract(pat = '([0-9]{4})').astype(int)
df
```

```python
df=df.sort_values(by=['gross'],inplace=False,ascending=False)
df
```

```python
sns.regplot(x="gross", y="budget", data=df, scatter_kws={"color":"red"},line_kws={"color":"blue"})
```

```python
df.corr(method="pearson")
```

```python
corr_matrix=df.corr(method="pearson")
sns.heatmap(corr_matrix,annot=True)
plt.title("Correlation Matrix for Numeric Features")
plt.xlabel("Movie Features")
plt.ylabel("Movie Featues")
plt.show()
```

```python
print('size of x = {0}'.format(df['gross'].size))
print('size of y = {0}'.format(df['budget'].size))
assert df['gross'].size == df['budget'].size 
```

```python
df_numerized=df

for col in df_numerized.columns:
    if (df_numerized[col].dtype=='object'):
        df_numerized[col]=df_numerized[col].astype('category')
        df_numerized[col]=df_numerized[col].cat.codes
        
df_numerized
```

```python
cor_mat=df_numerized.corr()
cor_pairs=cor_mat.unstack()
cor_pairs
```

```python
sorted_pairs=cor_pairs.sort_values()
sorted_pairs
```

```python
high_corr=sorted_pairs[(sorted_pairs)>0.5]
high_corr
```
